# **SOAL 3:** Makanan

```python
datapenjualan_makananpenutup = {
    'Donat': 14,
    'Pastel': 5,
    'Kue Coklat': 12,
    'Ice Cream': 19,
    'Puding Vanila': 6,
    'Brownies': 8,
    'Puding Strawberi': 12,
    'Puding Coklat': 9,
    'Ice Cream Nutela': 10,
    'Kue Coklat-Keju': 17,
    'Kue Wajik': 2,
    'Kue Sus': 9,
    'Mochi': 13,
}

MakPen = list(datapenjualan_makananpenutup .keys())
JumPen = list(datapenjualan_makananpenutup.values())

df = pd.DataFrame({
    'Nama': MakPen,
    'Penjualan': JumPen
})

df.sort_values(by='Penjualan', inplace=True)

colors = ['blue'] * len(df)
colors[0:2] = ['red'] * 3

plt.title ('Penjualan Makanan Penutup')
x_coords = np.arange(len(df))
plt.bar(
    x_coords,
    df['Penjualan'],
    tick_label = df['Nama'],
    color=colors
)

plt.xticks(rotation=90)
plt.xlabel('Paling sedikit ke Paling banyak')
plt.ylabel('Jumlah Penjualan')

plt.show()
```

Terlihat dari bagan bar penjualan makanan penjualan di atas yang menunjukan kepopuleran dari berbagai makanan penutup yang tersedia mulai dari Ice Cream yang paling populer dan terjual terbanyak, diikuti oleh Kue Coklat-Keju, dst. dan untuk pertanyaan sebutkan tiga makanan penutup yang harus disingkirkan?, dari bagan bar diatas terlihat bahwa makanan penutup Kue Wajik adalah yang paling sedikit terjual diikuti oleh, Pastel, dan Puding Vanila; Jadi, untuk tiga makanan penutup yang harus disingkirkan yaitu **Kue Wajik, Pastel, dan Puding Vanila**
